### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
print ("Object-oriented approach:\n")

import matplotlib.pyplot as plt
import numpy as np

# Datan
x = np.linspace(-4, 4, 100)
y = x**2


fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("y = x^2")
ax.set_xlabel("x")
ax.set_ylabel("y")
plt.show()

print ("Pyplot approach:\n")

# Datan
x = np.linspace(-4, 4, 100)
y = x**2

plt.plot(x, y)
plt.title("y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.show()

print ("Objekt approachen använder sig av fig och ax för mer kontroll, medans Pyplot är enkel och mer direkt. Bägge ger samma resultat")
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Data för scatter-plot
np.random.seed(15)
random_data_x = np.random.randn(1000)
random_data_y = np.random.randn(1000)
x = np.linspace(-2, 2, 100)
y = x**2

# Data för bar-plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())

# De två plottarna
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Scatter-plotten
ax1.scatter(random_data_x, random_data_y, alpha=0.5, color='orange', label='Scatter data')
ax1.plot(x, y, color='green', label='y = x^2')
ax1.set_title("Scatter Plot")
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.legend()

# Bar-plotten
ax2.bar(names, values, color='purple')
ax2.set_title("Bar Plot")
ax2.set_xlabel("Fruit")
ax2.set_ylabel("Quantity")

# Print av graferna
plt.tight_layout()
plt.show()
```
